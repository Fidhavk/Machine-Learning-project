```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
data = pd.read_csv('/content/student_habits_performance.csv')
```

```python
data
```

```python
data.shape
```

```python

data.info()
```

```python
data.isnull().sum()
```

```python
data['parental_education_level']=data['parental_education_level'].fillna('No')

data['parental_education_level']=data['parental_education_level'].replace({
    'None' : 'No',
    'Bachelor':'Yes',
    'Master':'Yes',
    'High School':'Yes'
})
data['parental_education_level']= data['parental_education_level'].map({'No':0,'Yes':1})
```

```python
data.isnull().sum()
```

```python
data['parental_education_level'].value_counts()
```

```python
data.duplicated().sum()
```

```python
data.columns
```

```python
data.drop('student_id',axis=1,inplace=True)
```

```python
le = LabelEncoder()
data['gender']=le.fit_transform(data['gender'])
data['part_time_job']=le.fit_transform(data['part_time_job'])
data['diet_quality']=le.fit_transform(data['diet_quality'])
data['parental_education_level']=le.fit_transform(data['parental_education_level'])
data['internet_quality']=le.fit_transform(data['internet_quality'])
data['extracurricular_participation']=le.fit_transform(data['extracurricular_participation'])
```

```python
data.info()
```

```python
x = data.drop('exam_score',axis=1)
y = data['exam_score']
```

```python
x
```

```python
y
```

#Linear Regression


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=30)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
pd.DataFrame({
    'actual': y_test,
   ' predict':y_pred
})
```

```python
model.score(x_train,y_train)*100
```

```python
model.score(x_test,y_test)*100
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python

r2_score(y_test,y_pred)*100
```

#Svm


```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python

x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.20,random_state=25)
```

```python
y_pred = model.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

#Descion Tree


```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=45)
```

```python
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python

r2_score(y_test,y_pred)
```

#Random Forest
